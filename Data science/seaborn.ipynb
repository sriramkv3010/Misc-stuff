```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
t=sns.load_dataset("tips")
t
```

```python
# Scatter Plot -  axis level function
# axis level function
sns.scatterplot(data=t,x="total_bill",y="tip",hue="sex",style="time",size="size")
```

```python
# figure level func - relplot, difference in height and width of plots -> in relplot its lightly squarish
# hue hidden parameter which classifies data similar to legend()
# style parameter does even more classifying
# size parameter replacement of annotations it increases size of each point based on quantity
sns.relplot(data=t,x="total_bill",y="tip",kind="scatter",hue="sex",style="time",size="size")
```

```python
# better to use relplot since it contains different kinds of axis level functions
```

```python
c= px.data.gapminder() # inbuilt dataset
c
```

```python
cn=c[c["country"] == "India"]
cn
```

```python
#axis level function
sns.lineplot(data=cn,x="year",y="lifeExp")
```

```python
# figure level function
sns.relplot(data=cn,x="year",y="lifeExp",kind="line")
```

```python
cn1=c[c["country"].isin(["India","Germany","Pakistan"])]
cn1
```

```python
sns.relplot(data=cn1,x="year",y="lifeExp",kind="line",hue="country",style="continent")
```

```python
# facet plot - plots grahs side by side based on category -> figure lvl function
sns.relplot(data=t,x="total_bill",y="tip",kind="scatter",col="smoker",row="sex")
```

```python
sns.relplot(data=t,x="total_bill",y="tip",kind="scatter",col="sex",row="day")
```

```python
# col wrap -> controls no of columns
sns.relplot(data=c,x="lifeExp",y="gdpPercap",kind="scatter",col="year",col_wrap=4)
```

```python
# DISTRIBUTIONAL PLOT - for 1 single variable
# univariate analysis
# figure lvl function -displot
# axis lvl function - histplot, kdeplot, rugplot
```

```python
t
```

```python
# histplot - axis lvl function
sns.histplot(data=t,x="total_bill")
```

```python
sns.displot(data=t,x="total_bill",kind="hist") # -> figure level function
```

```python
sns.displot(data=t,x="total_bill",kind="hist",bins=20)
```

```python
# It is also possible to plot a histogram on categorical data
# uses discrete bins
sns.displot(data=t,x="day",kind="hist")
# although it is basically a count plot
```

```python
sns.displot(data=t,x="tip",kind="hist",hue="sex")
```

```python
ts=sns.load_dataset("titanic")
ts
```

```python
sns.displot(data=ts,x="age",kind="hist",hue="sex")
```

```python
# facetplot - differentiates between given parameters
sns.displot(data=ts,x="age",kind="hist",col="sex")
```

```python
# kdeplot - kernal density estimation plot
# instead of using discrete bins, a kde plot smoothens the given data using gaussian kernel, and produces a continous density estimate
sns.kdeplot(data=t,x="total_bill")
# Here y axis represents probability of density
```

```python
sns.displot(data=t,x="total_bill",kind="kde")
```

```python
sns.displot(data=t,x="total_bill",kind="kde",hue="sex",fill=True)
```

```python
# Rugplot - plots marginal distribution by drawing ticks along with x and y axis
sns.rugplot(data=t,x="total_bill")
```

```python
# MATRIX PLOT - 1.heatmap 2.cluster map
# heatmap - grid of information
c
```

```python
nc = c.pivot(index="country",columns="year",values="lifeExp")
nc
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(nc)
```

```python
nc2=c[c["continent"]== "Europe"]
nc3=nc2.pivot(index="country",columns="year",values="lifeExp")
nc3
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(nc3)
```

```python
#annot - displays number along with color
plt.figure(figsize=(15,15))
sns.heatmap(nc3,annot=True,linewidth=0.5)
```
