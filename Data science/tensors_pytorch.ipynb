Inplace operations

```python
m = torch.rand(2,3)
n = torch.rand(2,3)
print(m)
print(n)
```

```python
m+n # a new tensor is being created and memory is allocated, in working with real dataset with large size too space consuming
```

```python
m.add_(n) # add_() - new tensor sum is being stored in m
```

```python
m
```

```python
# irrespective of function, at the end use _ so that instead of allocating memory to a new tensor it uses the old one
# inplace operator
```

```python
# copying a tensor - instead of using a assignment operator = which leads to allocation of same memory space better to use .clone()
```

```python
a=torch.rand(2,3)
a
```

```python
b=a.clone()
b # here memory location of a and b differs
```

```python
id(a)
```

```python
id(b)
```

```python
# moving a tensor from cpu to gpu, use .to()
b=a.to(d)
b
```
