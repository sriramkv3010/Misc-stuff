```python
#3D SURFACE PLOTS
# plotting a graph for f(x,y) = x**2+y**2
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y) #we get 2 arrays, meshgride maps 2 1D arrays, one to vertical and other to horizontal
xx
```

```python
yy
```

```python
z=xx**2+yy**2
z
```

```python
fig = plt.figure(figsize=(14,7))
ax = plt.subplot(projection="3d")
sp=ax.plot_surface(xx,yy,z,cmap="viridis")
plt.colorbar(sp)
```

```python
z=np.sin(xx)+np.cos(yy)
z
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection="3d")
sp = ax.plot_surface(xx,yy,z,cmap="viridis")
plt.colorbar(sp)
```

```python
z=np.sin(xx)+np.log(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection="3d")
sp = ax.plot_surface(xx,yy,z,cmap="viridis")
plt.colorbar(sp)
```

```python
z=np.sin(xx)+np.tan(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection="3d")
sp = ax.plot_surface(xx,yy,z,cmap="viridis")
plt.colorbar(sp)
```

```python
# CONTOUR PLOT - it represents 3d graph 2 dimensionally
```

```python
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
sp = ax.contour(xx,yy,z,cmap="viridis")
plt.colorbar(sp)
```

```python
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot()
sp = ax.contourf(xx,yy,z,cmap="viridis") #filled contour graph - contourf
plt.colorbar(sp)
```

```python
df = pd.read_csv("/content/IPL_Ball_by_Ball_2008_2022.csv")
df
```

```python
df.sample(10)
```

```python
ndf=df[(df["ballnumber"].isin([1,2,3,4,5,6])) & (df["batsman_run"]==6)]
ndf
```

```python
#pivot_table() in pandas it provides summary of a given dataframe
grid=ndf.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc="count")
grid
```

```python
plt.figure(figsize=(20,8))
plt.imshow(grid)
plt.colorbar()
```
