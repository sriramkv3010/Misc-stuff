````python
#broadcasting - describes how numpy operatesd on arrays with different sizes upon airthmetic operations
````

````python
#same shape
a=np.arange(6).reshape(3,2)
b=np.arange(6,12).reshape(3,2)
print(a)
print(b)
print(a+b)
````

````python
#different shape - broadcasting is done (smaller array is broadcasted across larger array)
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
````

````python
a=np.arange(12).reshape(4,3)
b=np.arange(3)
print(a)
print(b)
print(a+b)
````

````python
a=np.arange(3).reshape(3,1)
b=np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
````
